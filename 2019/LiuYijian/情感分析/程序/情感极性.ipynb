```python
path='/Users/lyj/Desktop/强度集合/8.txt'

f=open(path,'r+',encoding='utf-8')
lines=f.readlines()
f.close()

count_pos_list=[]
count_neg_list=[]
# count_neu_list=[]
# count_all_list=[]

print(len(lines))
for i in range(len(lines)):
    
    count_pos=0
    count_neg=0
#     count_neu=0
    
    line=(lines[i]).strip().split()
    for value in line:
        if float(value)<0.5:
            count_neg+=1
        else:
            count_pos+=1
#         else:
#             count_neu+=1
    count_pos_list.append(count_pos)
    count_neg_list.append(count_neg)
#     count_neu_list.append(count_neu)
#     count_all_list.append(len(line))

# print(count_all_list)
print(count_pos_list)
print(count_neg_list)
# print(count_neu_list)

```

```python
polarity=[]
polarity.append(sum(count_pos_list))
polarity.append(sum(count_neg_list))
#polarity.append(sum(count_neu_list))
polarity_list.append(polarity)
```
