```python
import os
import sys
import pandas as pd
import csv 
import datetime
import time
import numpy as np
```

```python
rootdir = 'C:/Users/Yee Shen/Desktop/source/cleandata/'
file_list = os.listdir(rootdir)


file_name = []
for i in range(0,len(file_list)):
    path = os.path.join(rootdir,file_list[i])
    if os.path.isfile(path):
        
        file_name.append(path)
        file_name = sorted(file_name)
```

```python
template = []
todel_name = file_name[10:19]
for delname in todel_name:
    template.append(file_name.pop(file_name.index(delname)))

file_name.extend(todel_name)
```

```python
def judge_repo(x):
    flag = False
    
    if ("抽" in x )and ("转" in x) :
        flag = True
    
    return flag
```

```python
def judge_comm(x):
    flag = False
    
    if ("抽" in x) and ("评" in x) :
        flag = True
    
    return flag
```

```python
def judge_like(x):
    flag = False
    
    if  ("抽" in x) and ("赞" in x) :
        flag = True
    
    return flag
```

```python
def judge_repo_comm(x):
    flag = False
    
    if  ("抽" in x) and( "转"in x) and ("评"in x) and ("赞" not in x ):
        flag = True
    
    return flag
```

```python
def judge_repo_like(x):
    flag = False
    
    if  ("抽" in x) and ("转"in x) and( "赞"in x) and ("评" not in x ):
        flag = True
    
    return flag
```

```python
def judge_comm_like(x):
    flag = False
    
    if  ("抽" in x) and ("评"in x ) and ("赞"in x) and ("转" not in x ):
        flag = True
    
    return flag
```

```python
def judge_repo_comm_like(x):
    flag = False
    
    if  "抽" in x and "评"in x and "赞"in x and "转"in x:
        flag = True
    
    return flag
```

```python
result = pd.DataFrame({'repo_percent':[],'comm_percent':[],'like_percent':[],
                       'repo_comm_percent':[],'repo_like_percent':[],'comm_like_percent':[],
                       'repo_comm_like_percent':[],'userid':[]})

for i in range(len(file_name)):
    
        mid_data = pd.read_csv(file_name[i])
        mid_data1 = mid_data[mid_data['text'].notnull()]
        
        if len(mid_data1) == 0:
            continue
        else:
            mid_data1['judge_repo'] = mid_data1['text'].map(lambda x: judge_repo(x))
            mid_data1['judge_comm'] = mid_data1['text'].map(lambda x: judge_comm(x))
            mid_data1['judge_like'] = mid_data1['text'].map(lambda x: judge_like(x))
            mid_data1['judge_repo_comm'] = mid_data1['text'].map(lambda x: judge_repo_comm(x))
            mid_data1['judge_repo_like'] = mid_data1['text'].map(lambda x: judge_repo_like(x))
            mid_data1['judge_comm_like'] = mid_data1['text'].map(lambda x: judge_comm_like(x))
            mid_data1['judge_repo_comm_like'] = mid_data1['text'].map(lambda x: judge_repo_comm_like(x))

            repo_data = mid_data1[mid_data1['judge_repo']]
            repo_data = repo_data.reset_index(inplace=False)
            repo_data = repo_data.drop('index',axis=1)

            comm_data = mid_data1[mid_data1['judge_comm']]
            comm_data = comm_data.reset_index(inplace=False)
            comm_data = comm_data.drop('index',axis=1)

            like_data = mid_data1[mid_data1['judge_like']]
            like_data = like_data.reset_index(inplace=False)
            like_data = like_data.drop('index',axis=1)

            repo_comm_data = mid_data1[mid_data1['judge_repo_comm']]
            repo_comm_data = repo_comm_data.reset_index(inplace=False)
            repo_comm_data = repo_comm_data.drop('index',axis=1)

            repo_like_data = mid_data1[mid_data1['judge_repo_like']]
            repo_like_data = repo_like_data.reset_index(inplace=False)
            repo_like_data = repo_like_data.drop('index',axis=1)

            comm_like_data = mid_data1[mid_data1['judge_comm_like']]
            comm_like_data = comm_like_data.reset_index(inplace=False)
            comm_like_data = comm_like_data.drop('index',axis=1)

            repo_comm_like_data = mid_data1[mid_data1['judge_repo_comm_like']]
            repo_comm_like_data = repo_comm_like_data.reset_index(inplace=False)
            repo_comm_like_data = repo_comm_like_data.drop('index',axis=1)
        
            repo_percent = len(repo_data)/len(mid_data1)
            comm_percent = len(comm_data)/len(mid_data1)
            like_percent = len(like_data)/len(mid_data1)
            repo_comm_percent = len(like_data)/len(mid_data1)
            repo_like_percent = len(like_data)/len(mid_data1)
            comm_like_percent = len(like_data)/len(mid_data1)
            repo_comm_like_percent = len(like_data)/len(mid_data1)
        
        
        tmp = pd.DataFrame({'repo_percent':[repo_percent],'comm_percent':[comm_percent],'like_percent':[like_percent],
                            'repo_comm_percent':[repo_comm_percent],'repo_like_percent':[repo_like_percent],'comm_like_percent':[comm_like_percent],
                            'repo_comm_like_percent':[repo_comm_like_percent],'userid':[i+1]})
        result = pd.concat([result,tmp],axis=0)

result
```
