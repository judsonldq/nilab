```python
import pandas as pd 
data = pd.read_csv("alters信息.csv",encoding='utf-8')
data_list = data['detail'].tolist()

result = []
for element in data_list:
    tmp = []
    line_data = element.strip().split('\n')
    tmp.append(line_data[0])
    for ele in line_data:
        ele = ele.strip()
        if ele.startswith('关注') or ele.startswith('粉丝') or ele.startswith('微博'):
            tmp.append(ele)
    result.append(tmp)

    
data['result'] = result
```

```python
data_clean = data[['result']]
data_clean
```

```python
data_clean.to_csv('alters_detail_4.csv',index=False,encoding='utf-8-sig')
```
