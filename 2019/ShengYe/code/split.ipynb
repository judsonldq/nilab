```python
import pandas as pd 
data = pd.read_csv("altersä¿¡æ¯.csv",encoding='utf-8')
data_list = data['detail'].tolist()

result = []
for element in data_list:
    tmp = []
    line_data = element.strip().split('\n')
    tmp.append(line_data[0])
    for ele in line_data:
        ele = ele.strip()
        if ele.startswith('å…³æ³¨') or ele.startswith('ç²‰ä¸') or ele.startswith('å¾®åš'):
            tmp.append(ele)
    result.append(tmp)

    
data['result'] = result
```
