```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import ensemble
from sklearn import cross_validation
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from sklearn.cross_validation import train_test_split
from sklearn import metrics
from astropy.units import Ybarn
from pylab import *
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

gongzuobu = pd.read_csv('features_lable_final.csv')
gongzuobu
```

```python
'product_comm_sent','neg_com_perc','ave_price','price_var','update','turnover_month_appr',
```

```python
X = gongzuobu[[
               'product_comm_sent','neg_com_perc','ave_price','price_var','update',
               'original_weibo','repost_weibo','like_weibo','fans','av_comment','av_like','av_repost','follow'
               ]]
y = gongzuobu[['sales_month']]
#X_train,X_test,y_train,y_test = train_test_split(X,y,random_state = 1)
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y, test_size=0.4, random_state=0)
forest = RandomForestRegressor(n_estimators = 100,criterion='mse',
                                      random_state=1,n_jobs=-1)
forest.fit(X_train, y_train)
y_train_pred = forest.predict(X_train)
y_test_pred = forest.predict(X_test)
```

```python
y_test_pred
```

```python
print('R^2 train: %.3f, test: %.3f' % (r2_score(y_train, y_train_pred),
                                   r2_score(y_test, y_test_pred)))
```

```python
importances = forest.feature_importances_
std = np.std([tree.feature_importances_ for tree in forest.estimators_],
             axis=0)
indices = np.argsort(importances)[::-1]

# Print the feature ranking
print("Feature ranking:")

for f in range(X.shape[1]):
    print("%d. feature %d (%f)" % (f + 1, indices[f], importances[indices[f]]))
```

```python
a=range(35)
ymajorLocator  = MultipleLocator(50000)
xmajorLocator  = MultipleLocator(1)
ax = subplot()
ax.yaxis.set_major_locator(ymajorLocator)
ax.xaxis.set_major_locator(xmajorLocator)
plt.plot(a,y_test,'b-')
plt.plot(a,y_test_pred,'g-')
plt.ylabel('sales_month')
plt.show()
```

```python
# Plot the feature importances of the forest
plt.figure()
plt.title("Feature importances")
plt.bar(range(X.shape[1]), importances[indices],
       color="r")
plt.xticks(range(X.shape[1]), indices)
plt.xlim([-1, X.shape[1]])
plt.show()
```
