```python
def fans_level(to_compute_list):
    level_1 = 0
    level_2 = 0
    level_3 = 0
    level_4 = 0
    level_5 = 0
    for i in to_compute_list:
        if i < 10000:
            level_1 += 1
        if (i >= 10000) and (i < 100000):
            level_2 += 1
        if (i >= 100000) and (i < 500000):
            level_3 += 1
        if (i >= 500000) and (i < 1000000):
            level_4 += 1
        if  i >= 1000000 :
            level_5 += 1
    fans_lv_list = [level_1,level_2,level_3,level_4,level_5]
    return fans_lv_list
```

```python
fans_level(fans_list)
```

```python
def follows_level(to_compute_list):
    lv_1 = 0
    lv_2 = 0
    lv_3 = 0
    lv_4 = 0
    for i in to_compute_list:
        if i < 100:
            lv_1 += 1
        if (i >= 100) and (i < 500):
            lv_2 += 1
        if (i >= 500) and (i < 1000):
            lv_3 += 1
        if  i >= 1000 :
            lv_4 += 1
    follows_lv_list = [lv_1,lv_2,lv_3,lv_4]
    return follows_lv_list
```

```python
follows_list = test['follow'].tolist()
follows_level(follows_list)
```

```python
def weibo_level(to_compute_list):
    f = 0
    g = 0
    h = 0
    k = 0
    for i in to_compute_list:
        if i < 100:
            f += 1
        if (i >= 100) and (i < 1000):
            g += 1
        if (i >= 1000) and (i < 5000):
            h += 1
        if  i >= 5000 :
            k += 1
    weibo_lv_list = [f,g,h,k]
    return weibo_lv_list
```

```python
weibo_list = test['weibo'].tolist()
weibo_level(weibo_list)[3]
```

```python
def contact_level(to_compute_list):
    lv1 = 0
    lv2 = 0
    lv3 = 0
    lv4 = 0
    lv5 = 0
    for i in to_compute_list:
        if i == 1 :
            lv1 += 1
        if i == 2 :
            lv2 += 1
        if i == 3 :
            lv3 += 1
        if  i == 4 :
            lv4 += 1
        if  i > 4 :
            lv5 += 1
    contact_lv_list = [lv1,lv2,lv3,lv4,lv5]
    return contact_lv_list
```

```python
def overlap_level(to_compute_list):
    a = 0
    b = 0
    c = 0
    d = 0
    e = 0
    for i in to_compute_list:
        if i == 1 :
            a += 1
        if (i > 1) and (i <= 10):
            b += 1
        if (i > 10) and (i <= 30):
            c += 1
        if (i > 30) and (i <= 50):
            d += 1
        if  i > 50 :
            e += 1
    overlap_lv_list = [a,b,c,d,e]
    return overlap_lv_list
```

```python
overlap_list = test['length'].tolist()
overlap_level(overlap_list)
```

```python
contact_list = test['contact'].tolist()
contact_level(contact_list)
```
