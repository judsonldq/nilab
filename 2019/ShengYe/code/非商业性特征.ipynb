```python
import os
import sys
import pandas as pd
import csv 
import datetime
import time
import numpy as np
```

```python
rootdir = 'C:/Users/Yee Shen/Desktop/source/cleandata/'
file_list = os.listdir(rootdir)


file_name = []
for i in range(0,len(file_list)):
    path = os.path.join(rootdir,file_list[i])
    if os.path.isfile(path):
        
        file_name.append(path)
        file_name = sorted(file_name)
```

```python
template = []
todel_name = file_name[10:19]
for delname in todel_name:
    template.append(file_name.pop(file_name.index(delname)))

file_name.extend(todel_name)
file_name
```

```python
def judge(x):
    flag = True
    
    if  "上新" in x or "新品" in x:
        flag = False
    
    return flag
```

```python
def compute_mean(to_compute_list):
    count = 0
    element_sum = 0
    for i in to_compute_list:
        if i==None:
            continue
        if type(i) == str:
            i = 0
        count += 1
        element_sum += i
    return element_sum/count
```

```python
def judge_prize(x):
    flag = True
    
    if ("抽" in x )and ("转" in x) :
        flag = False
    if ("抽" in x) and ("评" in x) :
        flag = False
    if  ("抽" in x) and ("赞" in x) :
        flag = False
        
    return flag
```

```python
mid_data = pd.read_csv(file_name[0])
mid_data1 = mid_data[mid_data['text'].notnull()]
mid_data1 = mid_data1.where(mid_data.notnull(), None)
mid_data1
```

```python
mid_data1['judge'] = mid_data1['text'].map(lambda x: judge(x))
want_data = mid_data1[mid_data1['judge']]
want_data = want_data.reset_index(inplace=False)
want_data = want_data.drop('index',axis=1)
want_data
```

```python
result = pd.DataFrame({'nonc':[],'nonc_avrepost':[],'nonc_avcomment':[],'nonc_avlike':[],'userid':[]})

for i in range(len(file_name)):
    
        mid_data = pd.read_csv(file_name[i])
        mid_data1 = mid_data[mid_data['text'].notnull()]
        mid_data1 = mid_data1.where(mid_data.notnull(), None)

        mid_data1['judge'] = mid_data1['text'].map(lambda x: judge(x))
        want_data = mid_data1[mid_data1['judge']]
        want_data = want_data.reset_index(inplace=False)
        want_data = want_data.drop('index',axis=1)
        
        want_data['judge1'] = want_data['text'].map(lambda x: judge_prize(x))
        final_data = want_data[want_data['judge1']]
        final_data = final_data.reset_index(inplace=False)
        final_data = final_data.drop('index',axis=1)
        
        nonc = len(final_data)
        
        nonc_avrepost = compute_mean(final_data['repost'].tolist())
        nonc_avcomment = compute_mean(final_data['comment'].tolist())
        nonc_avlike = compute_mean(final_data['like'].tolist())

        
        tmp = pd.DataFrame({'nonc':[nonc],'nonc_avrepost':[nonc_avrepost],'nonc_avcomment':[nonc_avcomment],'nonc_avlike':[nonc_avlike],'userid':[i+1]})
        result = pd.concat([result,tmp],axis=0)

result
```

```python
result.to_csv('noncommercial_feature.csv', index = 0, encoding='utf_8_sig')
```
