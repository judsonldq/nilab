```python
import os
import sys
import math
import numpy as np
import pandas as pd
import csv 
import matplotlib.pyplot as plt 
from matplotlib.ticker import MultipleLocator,FormatStrFormatter
% matplotlib inline
```

```python
result = pd.read_csv('overlap_result.csv', encoding='utf8')
result_multi11 = result[result['length']>10]
result_multi11
```

```python
result_multi11['value'] = result_multi11['value'].map(lambda x: eval(x))
value_list = result_multi11['value'].tolist()
result_list = []

for ele in value_list:
    for element in ele:
        result_list.append(element)

result_set = list(set(result_list))

result_dic = {}
for ele in result_set:
    result_dic[ele] = result_list.count(ele)
    
cele = []
count = []
for key in result_dic:
    cele.append(key)
    count.append(result_dic[key])

multi11_cele = pd.DataFrame({'celebrity':cele,'count':count})
multi11_cele
```

```python
multi11_cele.to_csv('multi11_count.csv',index=False)
```

```python
multi11_cele = multi11_cele.sort_values(['count'],ascending = False)
multi11_cele
```

```python
cele_list = pd.read_excel('电商网红名单.xlsx',encoding = 'utf8')
merge_multi11 = pd.merge(multi11_cele,cele_list,how='left')
merge_multi11
```

```python
features = pd.read_csv('features_lable_network1.csv',encoding = 'utf8')
fans_lable = features[['celebrity','fans','label']]
merge_multi11_features = pd.merge(merge_multi11,fans_lable,how='left',on='celebrity')
merge_multi11_features
```
