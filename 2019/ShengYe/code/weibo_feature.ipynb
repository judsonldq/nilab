```python
def compute_mean(to_compute_list):
    count = 0
    element_sum = 0
    for i in to_compute_list:
        if i==None:
            continue
        if type(i) == str:
            i = 0
        count += 1
        element_sum += i
    return element_sum/count
```

```python
result = pd.DataFrame({'userid':[],'av_repost':[],'av_comment':[],'av_like':[],'av_re_repost':[],'av_re_comment':[],'av_re_like':[]})
for i in range(len(file_name)):
    mid_data = pd.read_csv(file_name[i])
    mid_data = mid_data.where(mid_data.notnull(), None)
    av_repost = compute_mean(mid_data['repost'].tolist())
    av_comment = compute_mean(mid_data['comment'].tolist())
    av_like = compute_mean(mid_data['like'].tolist())
    av_re_repost = compute_mean(mid_data['re_repost'].tolist())
    av_re_comment = compute_mean(mid_data['re_comment'].tolist())
    av_re_like = compute_mean(mid_data['re_like'].tolist())
    
#     user_id = [i + 1]
    tmp = pd.DataFrame({'userid':[i+1],'av_repost':[av_repost],
                        'av_comment':[av_comment],'av_like':[av_like],
                        'av_re_repost':[av_re_repost],'av_re_comment':[av_re_comment],'av_re_like':[av_re_like]})
    
    result = pd.concat([result,tmp],axis=0)
```
