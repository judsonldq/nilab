```python
def exchange_time(x):
    flag = datetime.datetime.strptime(str(x.split(' ')[0]),'%m/%d/%Y').strftime('%Y-%m-%d')
    return flag
```

```python
def daynum(str1,str2):
    d1 = datetime.datetime.strptime(str1[0:10],"%Y-%m-%d")
    d2 = datetime.datetime.strptime(str2[0:10],"%Y-%m-%d")
    num = (d1-d2).days
    return num
```

```python
result_time = pd.DataFrame({'time':[],'interval':[]})

time_list = want_data['new_time'].tolist()
start_time = time_list[0]
interval = 0
for i in range(len(time_list)-1):
    tmp_time = daynum(time_list[i+1],time_list[i])
    if tmp_time <= 5:
        continue
    if tmp_time > 5 and tmp_time <= 365:
        interval = daynum(time_list[i+1],start_time)
        tmp_data = pd.DataFrame({'time':[start_time],'interval':[interval]})
        result_time = pd.concat([result_time,tmp_data],axis=0)
        start_time = time_list[i+1]
    if tmp_time > 365:
        start_time = time_list[i+1]

result_time
```

```python
for i in range(len(file_name)):
    if i == 73:
        result_time = pd.DataFrame({'time':[0],'interval':[0]})
    
    else:
        mid_data = pd.read_csv(file_name[i])
        mid_data = mid_data[mid_data['text'].notnull()]
        mid_data['judge'] = mid_data['text'].map(lambda x: judge(x))
        want_data = mid_data[mid_data['judge']]
        want_data = want_data.reset_index(inplace=False)
        want_data = want_data.drop('index',axis=1)
        want_data['new_time'] = want_data['time'].map(lambda x : exchange_time(x))
        want_data = want_data.sort_values('new_time')

        time_list = want_data['new_time'].tolist()
        
        result_time = pd.DataFrame({'time':[],'interval':[]})
        start_time = time_list[0]
        interval = 0
        for j in range(len(time_list)-1):
            tmp_time = daynum(time_list[j+1],time_list[j])
            if tmp_time <= 7:
                continue
            if tmp_time > 7 and tmp_time <= 90:
                interval = daynum(time_list[j+1],start_time)
                tmp_data = pd.DataFrame({'time':[start_time],'interval':[interval]})
                result_time = pd.concat([result_time,tmp_data],axis=0)
                start_time = time_list[j+1]
            if tmp_time > 90:
                start_time = time_list[j+1]

    result_time.to_csv(str(i+1)+'.csv', encoding='utf_8_sig',index=0)
```
