```python
result = pd.DataFrame({'userid':[],'start_y':[]})
for i in range(len(file_name)):
    test = pd.read_csv(file_name[i])
    test = test[test['time'].notnull()]
    test['time'] = test['time'].apply(str)
    test['new_time'] = test['time'].map(lambda x : exchange_time(x))
    test = test.sort_values('new_time')
    test = test.reset_index(drop=True)
    str1 = test['new_time'][0]
    date1 = datetime.datetime.strptime(str1[0:10],"%Y-%m-%d")
    start_y = date1.year
    tmp = pd.DataFrame({'userid':[i+1],'start_y':[start_y]})
    result = pd.concat([result,tmp],axis=0)

result
```
