```python
for i in range(len(file_name)):
    result = pd.read_csv(file_name[i])
    result1 = result[result['repost'].isnull()]
    result2 = result[result['repost'].notnull()]
    tmp1 = result1.groupby('re_user').count()
    tmp1_ = tmp1.reset_index()
    tmp2 = result2.groupby('re_user').count()
    tmp2_ = tmp2.reset_index()
    
    count1 = tmp1_[['re_user','time']]
    re_user_list = list(count1['re_user'])
    time_list = list(count1['time'])
    user_time_dict = {}
    for n in range(len(time_list)):
        user_time_dict[re_user_list[n]] = time_list[n]
    
    count2 = tmp2_[['re_user','time']]
    re_user_2_list = list(count2['re_user'])
    time_2_list = list(count2['time'])
    user_time_2_dict = {}
    for k in range(len(time_2_list)):
        user_time_2_dict[re_user_2_list[k]] = time_2_list[k]
    
    name_list = list(user_time_dict.keys())
    name_list.extend(list(user_time_2_dict.keys()))
    name_list = list(set(name_list))
    
    like_list = []
    trans_list = []
    for j in range(len(name_list)):
        if user_time_dict.get(name_list[j]) is not None:
            like_list.append(user_time_dict[name_list[j]])
        else:
            like_list.append(0)
        if user_time_2_dict.get(name_list[j]) is not None:
            trans_list.append(user_time_2_dict[name_list[j]])
        else:
            trans_list.append(0)
    
    merge = np.array([name_list,like_list,trans_list])
    merge = merge.T
    df = pd.DataFrame(merge,columns=['name','like','trans'])
    df.to_csv(str(i)+'.csv', index=0, encoding='utf_8_sig')
```
