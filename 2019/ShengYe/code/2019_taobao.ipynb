```python
def compute_sum(to_compute_list):
    element_sum = 0
    for i in to_compute_list:
        if i==None:
            continue
        if type(i) == str:
            i = 0
        element_sum += i
    return element_sum
```

```python
amount = test['amount'].tolist()
```

```python
compute_sum(amount)
```

```python
result = pd.DataFrame({'userid':[],'sales_2019':[]})
for i in range(len(file_name)):
    mid_data = pd.read_csv(file_name[i])
    amount = mid_data['amount'].tolist()
    sales = compute_sum(amount)
    
    tmp = pd.DataFrame({'userid':[file_name[i]],'sales_2019':[sales]})
    
    result = pd.concat([result,tmp],axis=0)

result
```
