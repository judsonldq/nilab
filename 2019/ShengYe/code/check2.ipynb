```python
import pickle
```

```python
with open('ego_dic.pk','rb') as f:
    data = pickle.load(f)
    
```

```python
import pandas as pd
```

```python
df = pd.read_csv('altersinfo_merge.csv')
```

```python
df = df[['content','length','name','follow','fans','weibo']]
```

```python
key_list = list(data.keys())[0:2]
```

```python
for key in key_list:
    empty_df = pd.DataFrame({'content':[],'length':[],'name':[],'follow':[],'fans':[],'weibo':[]})
    name_list = data[key]
    for name in name_list:
        tmp_data = df[df['name']==name]
        empty_df = pd.concat([empty_df,tmp_data],axis=0)
    empty_df.to_csv(str(key)+'.csv')
```
