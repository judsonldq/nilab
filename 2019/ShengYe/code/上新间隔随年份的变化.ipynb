```python
for i in range(len(file_name)):
    
    if i == 73:
        result = pd.DataFrame({'year':[0],'interval_ave':[0]})

    else:
        test = pd.read_csv(file_name[i])
        test['std_time']=pd.to_datetime(test.time,format='%Y/%m/%d')
        test['year'] = test['std_time'].map(lambda x : x.year)
        
        start_year = test['year'][0]
        year_list = test['year'].tolist()
        interval_sum = 0
        num = 0
        result = pd.DataFrame({'year':[],'interval_ave':[]})
        
        if len(year_list) == 1:
            tmp = pd.DataFrame({'year':[start_year],'interval_ave':[test['interval'][0]]})
            result = pd.concat([result,tmp],axis=0)
        
        else:
            for j in range(len(year_list)):

                if j < len(year_list) - 1:

                    if year_list[j] == start_year:
                        interval_sum += test['interval'][j]
                        num = num + 1

                    else:
                        tmp = pd.DataFrame({'year':[start_year],'interval_ave':[interval_sum/num]})
                        result = pd.concat([result,tmp],axis=0)
                        start_year = year_list[j]
                        interval_sum = test['interval'][j]
                        num = 1

                else: 
                    tmp = pd.DataFrame({'year':[start_year],'interval_ave':[interval_sum/num]})
                    result = pd.concat([result,tmp],axis=0)
            
    result.to_csv(str(i+1)+'.csv', encoding='utf_8_sig',index=0)
        
```
