```python
import xgboost as xgb
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.cross_validation import train_test_split

gongzuobu = pd.read_csv('C:/Users/Yee Shen/Desktop/source/total_21features.csv')
gongzuobu
```

```python
X = gongzuobu[[
               'product_comm_sent','neg_com_perc','ave_price','price_var','update','turnover_month_appr',
               'original_weibo','repost_weibo','like_weibo','fans','av_comment','av_like','av_repost','follow','vip_level'
               ]]
y = gongzuobu[['sales_month']]

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1000)
print(X_train.shape, X_test.shape)

#模型参数设置
xlf = xgb.XGBRegressor(max_depth=4, 
                       learning_rate=0.1, 
                       n_estimators=10, 
                       silent=True, 
                       objective='reg:linear', 
                       nthread=-1, 
                       gamma=0,
                       min_child_weight=1, 
                       max_delta_step=0, 
                       subsample=0.85, 
                       colsample_bytree=0.7, 
                       colsample_bylevel=1, 
                       reg_alpha=0, 
                       reg_lambda=1, 
                       scale_pos_weight=1, 
                       seed=1500, 
                       missing=None)

xlf.fit(X_train, y_train, eval_metric='rmse', verbose = True, eval_set = [(X_test, y_test)],early_stopping_rounds=100)

# 计算 auc 分数、预测
preds = xlf.predict(X_test)
```
