```python
import math
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import ensemble
from sklearn import cross_validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
from sklearn import preprocessing
from sklearn.preprocessing import PolynomialFeatures
from sklearn.grid_search import GridSearchCV
from sklearn.cross_validation import train_test_split
from astropy.units import Ybarn
from pylab import *
from matplotlib.ticker import MultipleLocator, FormatStrFormatter

gongzuobu = pd.read_csv('C:/Users/Yee Shen/Desktop/source/total_21features.csv')
gongzuobu
```

```python

def get_label(x):
    if x<43000 -75000/2:
        return 1
    if (x>= 43000 - 75000/2) and (x < 43000 + 75000/2):
        return 2
    if x>=43000 + 75000/2:
        return 3



gongzuobu['label'] = gongzuobu['sales_month'].map(lambda x : get_label(x))
set(gongzuobu['label'].tolist())
gongzuobu
```

```python
X = gongzuobu[[
               'product_comm_sent','neg_com_perc','ave_price','price_var','update','turnover_month_appr',
               'original_weibo','repost_weibo','like_weibo','fans','av_comment','av_like','av_repost','follow','vip_level'
               ]]
y = gongzuobu[['label']]
X_train, X_test, y_train, y_test = cross_validation.train_test_split(X,y, test_size=0.4, random_state=0)
rf0 = RandomForestClassifier(oob_score=True, random_state=10)  
rf0.fit(X,y)  
print (rf0.oob_score_)  
```

```python
init = rf0.predict_proba(X)

result = []

for ele in init:
    result.append(np.argmax(ele)+1)
```
