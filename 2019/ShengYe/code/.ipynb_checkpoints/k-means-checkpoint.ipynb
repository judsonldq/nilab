```python
import pandas as pd 
import matplotlib.pyplot as plt  
from sklearn.cluster import KMeans
from sklearn.externals import joblib
from sklearn.cluster import KMeans  
```

```python
alter = pd.read_csv('C:/Users/Yee Shen/Desktop/source/cleandata_alter/00.csv')
alter
```

```python
SSE = []
for k in range(3,10):  
    estimator = KMeans(n_clusters=k)  
    estimator.fit(alter[['contact']])  
    SSE.append(estimator.inertia_)  
X = range(3,10)  
plt.xlabel('k')  
plt.ylabel('SSE')  
plt.plot(X,SSE,'o-')  
plt.show()  
```
