```python
import os
import sys
import math
import numpy as np
import pandas as pd
import csv 
import matplotlib.pyplot as plt 
```

```python
df = pd.read_csv('alter_overlap.csv', encoding='utf_8_sig')
df.head()
```

```python
putong_wanghong_dict = {}

re_user_list = list(set(df['re_user'].tolist()))

user_id_set = list(set(df['userid'].tolist()))

for re_user in re_user_list:
    for user_id in user_id_set:
        tmp = df[df['re_user']==re_user]
        if len(tmp)!=0:
            tmp = tmp[tmp['userid']==user_id]
            if len(tmp)!=0:
                putong_wanghong_dict[re_user] = user_id
```

```python
putong_wanghong_dict = {}

re_user_list = list(set(df['re_user'].tolist()))

for re_user in re_user_list:
    putong_wanghong_dict[re_user] = []

user_id_set = list(set(df['userid'].tolist()))

for re_user in re_user_list:
    tmp = df[df['re_user'] == re_user] 
    if len(tmp) == 0:
        continue
    else:
        for user_id in user_id_set:
            tmp = tmp[tmp['userid']==user_id]
            if len(tmp)!=0:
                putong_wanghong_dict[re_user].append(user_id) 
```
