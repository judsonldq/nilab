```python
import pandas as pd
```

```python
data = pd.read_csv('overlap_result1.csv',encoding='utf-8')
```

```python
data
```

```python
data.iloc[100]
```

```python
name_list = []
content_list = []
count_list = []

for i in range(len(data)):
    tmp = data.iloc[i].tolist()
    name_list.append(tmp[0])  # 将名字追加上去
    
    
    last_data = tmp[1:]
    capacity = []
    for ele in last_data:
        if str(ele)=='nan':
            continue
        else:
            capacity.append(ele)
    count_list.append(capacity[-1])
    
    content_list.append(eval(''.join(capacity[0:-1])))
    
    
```

```python
content_list
```

```python
new_data = pd.DataFrame({'name':name_list,'content':content_list,'length':content_list})
```
