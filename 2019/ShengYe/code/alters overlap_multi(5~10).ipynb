```python
import os
import sys
import math
import numpy as np
import pandas as pd
import csv 
import matplotlib.pyplot as plt 
from matplotlib.ticker import MultipleLocator,FormatStrFormatter
% matplotlib inline
```

```python
result = pd.read_csv('overlap_result.csv', encoding='utf8')
result_multi5 = result[(result['length']>4) & (result['length']<11)]
result_multi5
```

```python
result_multi5['value'] = result_multi5['value'].map(lambda x: eval(x))
value_list = result_multi5['value'].tolist()
result_list = []

for ele in value_list:
    for element in ele:
        result_list.append(element)

result_set = list(set(result_list))

result_dic = {}
for ele in result_set:
    result_dic[ele] = result_list.count(ele)
    
cele = []
count = []
for key in result_dic:
    cele.append(key)
    count.append(result_dic[key])

multi5_cele = pd.DataFrame({'celebrity':cele,'count':count})
multi5_cele
```

```python
multi5_cele.to_csv('multi5_count.csv',index=False)
```

```python
multi5_cele = multi5_cele.sort_values(['count'],ascending = False)
multi5_cele
```

```python
cele_list = pd.read_excel('电商网红名单.xlsx',encoding = 'utf8')
merge_multi5 = pd.merge(multi5_cele,cele_list,how='left')
merge_multi5
```
