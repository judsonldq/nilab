```python
def judge(x):
    flag = False
    
    if  "上新" in x or "新品" in x:
        flag = True
    
    return flag
```

```python
def compute_mean(to_compute_list):
    count = 0
    element_sum = 0
    for i in to_compute_list:
        if i==None:
            continue
        if type(i) == str:
            i = 0
        count += 1
        element_sum += i
    return element_sum/count
```

```python
def exchange_time(x):
    flag = datetime.datetime.strptime(str(x.split(' ')[0]),'%m/%d/%Y').strftime('%Y-%m-%d')
    return flag
```

```python
result = pd.DataFrame({'userid':[],'new':[],'new_avrepost':[],'new_avcomment':[],'new_avlike':[]})
for i in range(len(file_name)):
    if i == 73:
        tmp = pd.DataFrame({'userid':[74],'new':[0],'new_avrepost':[0],'new_avcomment':[0],'new_avlike':[0]})
        result = pd.concat([result,tmp],axis=0)
    
    else:
        mid_data = pd.read_csv(file_name[i])
        mid_data = mid_data[mid_data['text'].notnull()]
        mid_data = mid_data.where(mid_data.notnull(), None)
        mid_data['judge'] = mid_data['text'].map(lambda x: judge(x))
        want_data = mid_data[mid_data['judge']]
        want_data = want_data.reset_index(inplace=False)
        want_data = want_data.drop('index',axis=1)
        want_data['new_time'] = want_data['time'].map(lambda x : exchange_time(x))
        want_data = want_data.sort_values('new_time')
        
        new = len(want_data)
        
        new_avrepost = compute_mean(want_data['repost'].tolist())
        new_avcomment = compute_mean(want_data['comment'].tolist())
        new_avlike = compute_mean(want_data['like'].tolist())

        tmp = pd.DataFrame({'userid':[i+1],'new':[new],'new_avrepost':[new_avrepost],'new_avcomment':[new_avcomment],'new_avlike':[new_avlike]})
        result = pd.concat([result,tmp],axis=0)

result
       
```
